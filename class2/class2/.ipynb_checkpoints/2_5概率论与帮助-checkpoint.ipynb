```python
%matplotlib inline
import torch
from torch.distributions import multinomial
from d2l import torch as d2l
```

### 创立一个投色子模型，首先建立一个概率向量，在索引i处的值是采样结果中i出现的次数

```python
fair_probs = torch.ones([6]) / 6
multinomial.Multinomial(1, fair_probs).sample(), fair_probs
```

### 进行多次投掷，使用深度学习框架同时抽取多个样本(用for会很慢）

```python
counts = multinomial.Multinomial(10000, fair_probs).sample()
counts / 10000  # 每次单独投掷的概率
```

### 进行500组实验，每组抽10个样本
### counts得到500个向量，每个向量表示抽取10次各个数出现的次数
### count.cumsum(dim=0)表示本行与沿着列与上行相加，如第三行为3，为1+2,但保留所有向量
### cum_counts.sum(dim=1, keepdims=True)沿着行将向量元素相加，因为在counts中每个进行了10次抽样，所以每个向量中所有元素沿行相加为10，count.cunsum(dim=0)表示本行与上行相加，所以在第二行个元素沿行相加为10+10，同理第i行为10*i。keepdims=True表示原本沿行降维应为500，保持维度为(500,1)
### 所以最终estimate表示cum_counts的每一个向量除以对应的抽样次数。其中cum_counts为(500，6),cum_counts.sum为（500，1），故广播机制扩成500，6然后逐个元素对应相除

```python
counts = multinomial.Multinomial(10, fair_probs).sample((500,))
cum_counts = counts.cumsum(dim=0)
estimates = cum_counts / cum_counts.sum(dim=1, keepdims=True)
estimates
```

```python
for i in range(6):
    print(estimates[:, i].shape)
```

### estimates[:, i]表示取estimates左右行的第i列元素，新组成的是一个500维向量

```python
d2l.set_figsize((6, 4.5))
for i in range(6):
    d2l.plt.plot(estimates[:, i].numpy(),
                label=("P(die=" + str(i + 1) + ")"))
d2l.plt.axhline(y=0.167, color='black', linestyle='dashed')
d2l.plt.gca().set_xlabel('Groups of experiments')
d2l.plt.gca().set_ylabel('Estimated probability')
d2l.plt.legend();
```
